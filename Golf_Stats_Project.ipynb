### Import needed modules and files for project

```python
import pandas as pd

golf_data = pd.read_csv('C:\\Users\\tarellanes\\Downloads\\PGATOUR_data2 (2).csv')
```

### Sorting CSV file by points in ascending order

```python
golf_data = golf_data.sort_values('POINTS', ascending=False)
golf_data
```

### Displaying the top 25 golfers by points

```python
sorted_25_golf = golf_data.head(25)
sorted_25_golf
```

### Further sorting data this time by columns of interest using filter function

```python
sorted_columns = sorted_25_golf.filter(['Player', 'POINTS', 'NUMBER_OF_WINS', 'AVG_driving_DISTANCE', 'UP_AND_DOWN_%', 'MISSED_GIR', 'FAIRWAY_HIT_%', 
                                        'Three_PUTT%', 'AVG_CLUB_HEAD_SPEED', 'AVG_SPIN_RATE', 'AVG_SCORE', 'AGE'])
sorted_columns
```

### Created a histogram to help visualize data 
### Used describe method to see general statistics around selected golf statistics

```python
sorted_columns.hist()
sorted_columns.describe().round(2)
```

### Using Correlation method to create matrix of data to be able to draw conclusions from

```python
final_golf_analysis = sorted_columns.corr().round(3)
final_golf_analysis
```

### Task 5 Bonus - Using Wordcloud

```python
from wordcloud import WordCloud, STOPWORDS
import wikipedia
from PIL import Image

stop_w = set(STOPWORDS)
info = wikipedia.summary("Golfing")
word_cloud = WordCloud(stopwords = stop_w).generate(info)
img = word_cloud.to_image()
img.show()
```
